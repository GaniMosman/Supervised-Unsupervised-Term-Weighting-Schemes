```python
import math
import pandas as pd
from ast import literal_eval

pd.set_option('display.max_columns', 2000)
pd.set_option('mode.use_inf_as_na', True)
```

## Reading the dataset

```python
preprcessed_question = pd.read_excel("../preprocessing_result/preprocessing_result.xlsx")
preprcessed_question.Preprocessed_Question = preprcessed_question.Preprocessed_Question.apply(literal_eval)
preprcessed_question
```

## Bag of Words creation

```python
#Function to create vocabulary for the BOW
def create_vocabulary(questions):
    unique_words = []
    for question in questions['Preprocessed_Question']:
        for word, tag in question:
            if word not in unique_words:
                unique_words.append(word)
        
    return unique_words

vocabulary = create_vocabulary(preprcessed_question)           
print(len(vocabulary))
print(vocabulary)
```

```python
#function to count word frequency of each word in a question
def count_word_frequency(question):
    bag_of_words_count = dict.fromkeys(vocabulary, 0)
    for word,_ in question:
        bag_of_words_count[word] += 1
    
    return bag_of_words_count

list_of_bow = []
for question in preprcessed_question['Preprocessed_Question'].values:
    bow = count_word_frequency(question)
    list_of_bow.append(bow)
```

```python
bow_df = pd.DataFrame(list_of_bow)
bow_df
```

# TFPOS-IDF

```python
def compute_total_POS_weight_with_freq(question, bow):
    
    weight = []
    for keyword, pos in question:
        
        if pos == 'v':
            weight.append(bow[keyword] * 5)
        elif pos == 'n' or pos == 'a':
            weight.append(bow[keyword] * 3)
        else:
            weight.append(bow[keyword] * 1)
            
    return sum(weight)
```

```python
def return_POS_weight(word, question):
    
    pos = ''
    for keyword, tag in question:
        if keyword == word:
            pos = tag
            break
    if pos == 'v':
        return 5
    elif pos == 'n' or pos == 'a':
        return 3
    else:
        return 1
```

```python
#Function to compute TF
def compute_term_frequency(bow, question):
    
    term_freq = {}
    total_POS_weight_freq = compute_total_POS_weight_with_freq(question, bow)
    
    for word, freq in bow.items():
        weight = return_POS_weight(word, question)
        term_freq[word] = (freq * weight) / (total_POS_weight_freq)
        
    return term_freq

x = 0
list_term_freq = []
for question in preprcessed_question['Preprocessed_Question'].values:
    freq = compute_term_frequency(list_of_bow[x], question)
    list_term_freq.append(freq)
    x += 1
```

```python
bow_tf = pd.DataFrame(list_term_freq)
bow_tf
```

```python
#Function to compute IDF
def compute_inv_frequency(bow, keywords):
    inv_freq = {}
    len_documents = bow.shape[0]
    for keyword in keywords:
        total_doc_contain_words = (bow[keyword] != 0).sum()
        print(keyword, total_doc_contain_words)
        idf = math.log10(len_documents / total_doc_contain_words) + 1
        inv_freq[keyword] = idf
        
    return inv_freq
        
inv_doc_freq = compute_inv_frequency(bow_df, vocabulary)   
```

```python
print(inv_doc_freq)
```

```python
bow_inv = pd.DataFrame([inv_doc_freq])
bow_inv
```

```python
#Function to compute TF-IDF
def compute_tf_idf(tf, idf):
    tf_idf_temp = {}
    for word, value in tf.items():
        tf_idf_temp[word] = value * idf[word]
    
    return tf_idf_temp


list_tf_idf = [] 
for tf in list_term_freq:
    tf_idf = compute_tf_idf(tf, inv_doc_freq)
    list_tf_idf.append(tf_idf)
```

```python
tf_idf_frame = pd.DataFrame(list_tf_idf)
tf_idf_frame['BT_level'] = preprcessed_question['BT LEVEL'].values
tf_idf_frame
```
