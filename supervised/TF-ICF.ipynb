```python
import math
import pandas as pd
from ast import literal_eval

pd.set_option('display.max_columns', 2000)
pd.set_option('mode.use_inf_as_na', True)
```

## Reading the dataset

```python
preprcessed_question = pd.read_excel("../preprocessing_result/preprocessing_result.xlsx")
preprcessed_question.Preprocessed_Question = preprcessed_question.Preprocessed_Question.apply(literal_eval)
preprcessed_question
```

```python
for question in preprcessed_question['Preprocessed_Question']:
    print(question)
```

## Bag of Words creation

```python
#Function to create vocabulary for the BOW
def create_vocabulary(questions):
    unique_words = []
    for question in questions['Preprocessed_Question']:
        for word, tag in question:
            if word not in unique_words:
                unique_words.append(word)
        
    return unique_words

vocabulary = create_vocabulary(preprcessed_question)           
print(len(vocabulary))
print(vocabulary)
```

# TF

```python
#Function to compute TF
def compute_term_frequency(question):
    bag_of_words_count = dict.fromkeys(vocabulary, 0)
    for word,_ in question:
        bag_of_words_count[word] += 1
    
    return bag_of_words_count

list_of_tf = []
for question in preprcessed_question['Preprocessed_Question'].values:
    bow = compute_term_frequency(question)
    list_of_tf.append(bow)
```

```python
bow_tf = pd.DataFrame(list_of_tf)
bow_tf['BT_level'] = preprcessed_question['BT LEVEL'].values
bow_tf
```

# ICF

```python
#Function to compute ICF
def compute_category_frequency(bow, vocabulary):
    
    inv_category_freq = {}
    classes = bow['BT_level'].value_counts().index.to_list()
    
    for term in vocabulary:
        num_of_class = 0
        
        for x in range(len(classes)):
            bow_df_class = bow[bow['BT_level'] == classes[x]]
            total_doc_contain_words = (bow_df_class[term] != 0).sum()
            
            print(term, total_doc_contain_words)
         
            if total_doc_contain_words > 0:
                num_of_class += 1
        
        icf = math.log10(1 + (len(classes) / num_of_class))
        inv_category_freq[term] = icf
        
    return inv_category_freq
        
inv_ct_freq = compute_category_frequency(bow_tf, vocabulary) 
```

```python
print(inv_ct_freq)
```

```python
bow_inv = pd.DataFrame([inv_ct_freq])
bow_inv
```

```python
#Function to compute TF-IDF
def compute_tf_icf(tf, icf):
    tf_icf_temp = {}
    for word, value in tf.items():
         tf_icf_temp[word] = value * icf[word]
    
    return tf_icf_temp


list_tf_icf = [] 
for tf in list_of_tf:
    tf_icf = compute_tf_icf(tf, inv_ct_freq)
    list_tf_icf.append(tf_icf)
```

```python
tf_icf_frame = pd.DataFrame(list_tf_icf)
tf_icf_frame['BT_level'] = preprcessed_question['BT LEVEL'].values
tf_icf_frame
```

```python
#Saving result to file

tf_icf_frame.to_csv('term_weighting_result_supervised/tf-icf.csv', index = False)
tf_icf_frame.to_excel('term_weighting_result_supervised/tf-icf.xlsx', index = False)
```
